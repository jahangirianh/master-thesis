```python
K = nx.read_edgelist("data/com-dblp.ungraph.txt", create_using=nx.Graph(), nodetype=int)
```

```python
print(nx.info(K))
```

```python
def structural_similarity(G, node1, node2):
    """

    """
    c_neighbours = len(list(nx.common_neighbors(G, node1, node2))) + 2

    n1_neighbours = len(G[node1]) + 1
    n2_neighbours = len(G[node2]) + 1
    # print(c_neighbours)

    ss = c_neighbours / ((n1_neighbours * n2_neighbours) ** 0.5)
    return ss
```

```python
def eps_neighbours(G, node, eps):
    en = set()
    n = set()
    en.add(node)
    n.add(node)
    for neighbour in G.neighbors(node):
        n.add(neighbour)
        if(structural_similarity(G, node, neighbour) > eps):
            en.add(neighbour)
    return en
```

```python
def check_core(G, node, eps, mu):
    en = eps_neighbours(G, node, eps)
    if(len(en) >= mu):
        return True
    else:
        return False
```

```python
def dirReach(G, node, eps, mu):
    dir = set()
    if(check_core(G, node, eps, mu)):
        dir = eps_neighbours(G, node, eps)

    return dir
```

```python
def SCAN(G, eps, mu):
    print(nx.get_node_attributes(G, 'label'))
    nx.set_node_attributes(G, 'not', 'label')
    nx.get_node_attributes(G, 'label')
    cluster_id = 0
    for n in G.nodes:
        if(check_core(G, n, eps, mu)):
            # print('----------')
            # print(n)
            cluster_id += 1
            quee = eps_neighbours(G, n, eps)
            while quee:
                y = quee.pop()
                R = dirReach(G, y, eps, mu)
                # print('q', quee)
                # print(R)
                # print('+++++++')
                while R:
                    r = R.pop()
                    label = G.nodes[r]['label']
                    if(label == 'not' or label == 'non-member'):
                        G.nodes[r]['label'] = cluster_id
                        # print(nx.get_node_attributes(G, 'label'))
                    if(label == 'not'):
                        # print(r)
                        quee.add(r)
        else:
            if(G.nodes[n]['label'] == 'not'):
                G.nodes[n]['label'] = 'non-member' 

    for n in G.nodes:
        if(G.nodes[n]['label'] == 'non-member'):
            if(len(G[n]) > 1):
                G.nodes[n]['label'] = 'Hub'

            else:
                G.nodes[n]['label'] = 'Hub' 

                
    return nx.get_node_attributes(G, 'label')
     

    
```

```python
print(SCAN(K, .7, 4))
```

```python
H = nx.read_gml("data/football/football.gml")
print(SCAN(H, .5, 2))
```
